# Feature Engineering


```
Feminine = ['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','gender']

Masculine = ['M1','M2','M3','M4','M5','M6','M7','M8','M9','M10','gender']


# create a new column of the average F/M
# loc's args are all rows and then specified columns
df_gender['avg_feminine'] = df_gender.loc[:,Feminine].sum(axis=1) /10
df_gender['avg_masculine'] = df_gender.loc[:,Masculine].sum(axis=1) /10
```

```
def binned_gender_crosstab(data, normalize=False, x='avg_feminine', y='avg_masculine', bins=[0,1,2,3,4,5,6]):
  return pd.crosstab(pd.cut(data[x], bins), pd.cut(data[y], bins, normalize))

binned_gender_crosstab(df_gender)
```
